```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("farmacie.sicilia.csv", sep=",")
```

```python
print(df.head())  # Mostra le prime righe
print(df.info())  # Informazioni sulle colonne
```

```python
farmacia_per_provincia = df['PROVINCIA'].value_counts()
# Crea un grafico a barre utilizzando plt.bar()

plt.figure(figsize=(10, 6))
plt.bar(farmacia_per_provincia.index, farmacia_per_provincia.values, color='skyblue')

# Aggiungi il titolo e le etichette degli assi
plt.title("Numero di farmacie per provincia in Sicilia")
plt.xlabel("Provincia")
plt.ylabel("Numero di farmacie")

# Ruota le etichette sull'asse x per una migliore visibilità
plt.xticks(rotation=90)

# Mostra il grafico
plt.show()

```

```python
# Filtra il DataFrame per la provincia di Agrigento
df_agrigento = df[df['PROVINCIA'] == 'Agrigento']

# Conta il numero di farmacie per comune nella provincia di Agrigento
farmacia_per_comune_agrigento = df_agrigento['COMUNE'].value_counts()

# Crea il grafico a barre
plt.figure(figsize=(10, 6))
plt.bar(farmacia_per_comune_agrigento.index, farmacia_per_comune_agrigento.values, color='skyblue')

# Aggiungi il titolo e le etichette degli assi
plt.title("Distribuzione delle farmacie nella provincia di Agrigento")
plt.xlabel("Comune")
plt.ylabel("Numero di farmacie")

# Ruota le etichette sull'asse x per una migliore visibilità
plt.xticks(rotation=90)

# Mostra il grafico
plt.show()

```
